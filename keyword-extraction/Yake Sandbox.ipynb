```python
import yake
```

```python
# basic yake params
kw_extractor = yake.KeywordExtractor()
language = "en"
max_ngram_size = 3
deduplication_threshold = 0.9
numOfKeywords = 20
```

```python
custom_kw_extractor = yake.KeywordExtractor(lan=language, n=max_ngram_size, dedupLim=deduplication_threshold, top=numOfKeywords, features=None)
```

```python
keywords_list=[]
```

```python
# open mdx file, run default yake keyword extractor, add keywords to keywords_list
with open('test_file.mdx', 'r') as file:
    text = file.read().replace('\n', '')
    keywords = custom_kw_extractor.extract_keywords(text)
    for kw in keywords:
        keywords_list.append(kw[0])
print(keywords_list)
```

```python
# open same mdx file, add all headings (#...) to keywords list
with open('test_file.mdx', 'r') as file:
    for line in file:
        if line.strip() != "":
            if line.strip()[0]=="#":
                keywords_list.append(line)
print(keywords_list)
```

```python
# open same mdx file, try to identify all software names mentioned in doc and add them to keywords list
with open('test_file.mdx', 'r') as file:
    for line in file:
        if "apt install" in line:
            start_index = line.find("apt install")
            end_index = start_index + len("apt install")
            restofline_tokens=line[end_index:].split()
            keywords_list.append(restofline_tokens[0])
        elif "apt-get install" in line:
            start_index = line.find("apt-get install")
            end_index = start_index + len("apt-get install")
            restofline_tokens=line[end_index:].split()
            keywords_list.append(restofline_tokens[0])
        elif any(x in line for x in ["install", "Install"]):
            start_index = line.find("nstall")
            end_index = start_index + len("nstall")
            restofline_tokens=line[end_index:].split()
            if restofline_tokens[0] in ["the", "a", "an", "your", "ing", "ed", "ation"]:
                restofline_tokens=restofline_tokens[1:]
            if restofline_tokens[0] in ["the", "a", "an", "your", "ing", "ed", "ation"]:
                restofline_tokens=restofline_tokens[1:]
            if restofline_tokens[0][0].isupper():
                keywords_list.append(restofline_tokens[0])
print(keywords_list)
```

```python
# remove duplicate keywords from list
keywords_list=(list(set(keywords_list)))
print(keywords_list)
```

```python
new_keywords_list=[]

# clean up keywords list
for keyword in keywords_list:
    if " " in keyword:
        keyword_tokens=keyword.split()
        if "#" in keyword_tokens[0]:
            keyword_tokens=keyword_tokens[1:]
        keyword = '-'.join(keyword_tokens)
        new_keywords_list.append(keyword)
    else:
        new_keywords_list.append(keyword)
```

```python
# sort keywords by length
sorted_list = sorted(new_keywords_list, key=len)
sorted_list = [x for x in sorted_list if x not in ["Lightbox-src", "HowToStep-url","HowToStep", "Lightbox"]]
print(sorted_list)
```
